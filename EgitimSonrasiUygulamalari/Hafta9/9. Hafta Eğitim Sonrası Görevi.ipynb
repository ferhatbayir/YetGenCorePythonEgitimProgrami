# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import time
import random

class Kumanda:
    def __init__(self):
        self.tv_durum = "kapalı"
        self.tv_ses = 0
        self.tv_kanal_listesi = []
        self.tv_kanal = ""

    def tv_ac(self):
        if self.tv_durum == "açık":
            print("Televizyonunuz açık")
        else:
            self.tv_durum = "açık"
            print("Televizyonunuz açılıyor")

    def tv_kapat(self):
        if self.tv_durum == "kapalı":
            print("Televizyonunuz kapalı")
        else:
            self.tv_durum = "kapalı"
            print("Televizyonunuz kapatılıyor")
        
    def ses_ayarlari(self):
        while True:
            cevap = input("Sesi açmak için '+', kısmak için'-', çıkmak için 'w' tuşuna basınız ")
            if cevap == "+":
                if self.tv_ses != 50:
                    self.tv_ses += 1
                    print("ses: ", self.tv_ses)
            elif cevap == "-":
                if self.tv_ses != 0:
                    self.tv_ses -= 1
                    print("ses: ", self.tv_ses)
            else:
                print("Ses Güncellendi: ", self.tv_ses)
                break

    def kanal_ekle(self, kanal_ismi):
        print("kanal ekleniyor")
        time.sleep(1)
        self.tv_kanal_listesi.append(kanal_ismi)
        print("kanal eklendi")

    def kanal_sayisi(self):
        print("Toplam", len(self.tv_kanal_listesi), "kanal var")

    def rastgele_kanal(self):
        rastgele = random.randint(0, len(self.tv_kanal_listesi)-1)
        self.tv_kanal = self. tv_kanal_listesi[rastgele]
        print("Şu anki kanal: ", self.tv_kanal)

    def tv_bilgileri(self):
        print("""
        Televizyon Bilgileri:
        TV Durumu: {}
        Ses Düzeyi: {}
        Kanal Listesi: {}
        Şu Anki Kanal:{}
""".format(self.tv_durum, self.tv_ses, self.tv_kanal_listesi, self.tv_kanal))
        
kumanda = Kumanda()

print("""
Televizyon Kumandası Uygulaması

1. TV Aç

2. TV Kapat

3. Ses Ayarları

4. Kanal Ekle

5. Kanal Sayısını Öğrenme

6. Rastgele Kanala Geçme

7. Televizyon Bilgileri

çıkmak için 'q' tuşuna basın
""")

while True:
    islem = input("işlemi seçiniz: ")

    if islem == "q":
        print("program sonlandırılıyor")
        break
    elif islem == "1":
        kumanda.tv_ac()
    elif islem == "2":
        kumanda.tv_kapat()
    elif islem == "3":
        kumanda.ses_ayarlari()
    elif islem == "4":
        kanal_isimleri = input("Eklemek istediğiniz kanalı ',' ile ayırarak giriniz.")
        kanal_listesi = kanal_isimleri.split(",")
        for kanal in kanal_listesi:
            kumanda.kanal_ekle(kanal)
    elif islem == "5":
        kumanda.kanal_sayisi()
    elif islem == "6":
        kumanda.rastgele_kanal()
    elif islem == "7":
        kumanda.tv_bilgileri()
    else:
        print("Geçersiz işlem")
```
